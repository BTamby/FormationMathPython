## Pour aller plus loin : Librairie **mathplolib** - Représentations graphiques de données

La librairie **mathplotlib** propose le module *pyplot*, entre autres, pour la représentation graphique de données.<br/>
La première chose à faire pour utiliser le module *pyplot* est de l'importer à partir de **matplotlib**.

```python
from matplotlib import pyplot as plt
```

### A) Un graphique simple

```python
x = [-15,10,20,30]
y = [8,0,4,9]
x2 = [-5,-2,6]
y2 = [-2,0,3]

# Les tracés : Ligne
plt.plot(x,y, label ='Série 1')
plt.plot(x2,y2, label ='Série 2')
# Les tracés : point
plt.scatter(x,y, marker ='o', s = 50)
plt.scatter(x2,y2, marker ='+', s = 100, color ='green')
# Informations diverses
plt.title('Représentation de données')
plt.ylabel('Ordonnées')
plt.xlabel('Abscisses')
plt.legend()
# Affichage
plt.show()
```

### B) Personnaliser la représentation des données

<p>Pour en savoir plus sur le module **pyplot** voir la documentation en ligne : [Documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot).</p>

#### 1) Un peu de mise en forme

```python
# Les données
x = [-15,10,20,30]
y = [8,0,4,9]
x2 = [-5,-2,6]
y2 = [-2,0,9]

# Les tracés
plt.plot(x,y, label ='Série 1', linestyle = '-.',linewidth =2, color ='#000000')
plt.plot(x2,y2, label ='Série 2', linewidth =4)

# Informations diverses
plt.title('Représentation de données', fontname ='Times new roman', size = 20, color ='red')
plt.ylabel('Ordonnées', fontname ='Arial', size = 12, fontweight = 'bold', color = '#0012ff')
plt.xlabel('Abscisses')
plt.legend(loc='lower center',)

# Affichage
plt.show() # Afficher le repère
```

#### 2) Centrer les axes

<p>On peut personnaliser le repère en donnant des valeurs particulières aux paramètres de la méthode *gca(). Pour connaître tous les parammètres, voir la documentation en ligne : [Documentation](https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure.gca).</p>

```python
x = [-15,10,20,30]
y = [8,0,4,9]
x2 = [-5,-2,6]
y2 = [-2,0,9]
# Les tracés : Ligne
plt.plot(x,y, label ='Série 1')
plt.plot(x2,y2, label ='Série 2')
# Les tracés : Point
plt.scatter(x,y, marker ='o', s = 50)
plt.scatter(x2,y2, marker ='+', s = 100, color ='green')
# Informations diverses
plt.title('Représentation de données', fontname ='Times new roman', size = 16, color ='red')
plt.ylabel('Ordonnées', fontname ='Arial', size = 12, fontweight = 'bold', color = '#0012ff')
plt.xlabel('Abscisses')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.1), shadow=True, fontsize= 12, facecolor ='#12ffff' )
# Centrer les axes
rep = plt.gca()# On récupère le repère dans lequel on a représenté les données
rep.spines['right'].set_visible(False) # Ne pas afficher la verticale à droite
rep.spines['top'].set_visible(False) # Ne pas afficher l'horizontale en haut
rep.spines['bottom'].set_position('zero') # Déplacer l'horizontale du bas en y = 0
rep.spines['left'].set_position('zero') # Déplacer la verticale de gauche en x = 0
rep.yaxis.set_ticks_position('left') # N'afficher les graduations que sur la verticale de gauche
rep.xaxis.set_ticks_position('bottom') # N'afficher les graduations que sur l'horizontale du bas
# Afficher la grille
plt.grid(linestyle = '--')
# Affichage
plt.show() 
```

#### 3) Configurer les graduations

```python
x = [-15,10,20,30]
y = [8,0,4,9]
x2 = [-5,-2,6]
y2 = [-2,0,9]
# Les tracés : Ligne
plt.plot(x,y, label ='Série 1')
plt.plot(x2,y2, label ='Série 2')
# Les tracés : Point
plt.scatter(x,y, marker ='o', s = 50)
plt.scatter(x2,y2, marker ='+', s = 100, color ='green')
# Informations diverses
plt.title('Représentation de données', fontname ='Times new roman', size = 16, color ='red')
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.1), shadow=True, fontsize= 12, facecolor ='#12ffff' )
# Centrer les axes
rep = plt.gca()# On récupère le repère dans lequel on a représenté les données
rep.spines['right'].set_visible(False) # Ne pas afficher la verticale à droite
rep.spines['top'].set_visible(False) # Ne pas afficher l'horizontale en haut
rep.spines['bottom'].set_position('zero') # Déplacer l'horizontale du bas en y = 0
rep.spines['left'].set_position('zero') # Déplacer la verticale de gauche en x = 0
rep.yaxis.set_ticks_position('left') # N'afficher les graduations que sur la verticale de gauche
rep.xaxis.set_ticks_position('bottom') # N'afficher les graduations que sur l'horizontale du bas
# Configurer l'affichage des unités sur l'axe des abscisses
rep.xaxis.set_major_formatter(plt.FuncFormatter('{:.2f}'.format))
# Afficher la grille
plt.grid(linestyle = '--')
# Affichage
plt.show() 
```

#### 4) Affichage LaTeX

<p>On peut si on le souhaite utiliser **LaTeX** pour afficher des informations sur le graphique. On utilisera les modules suivants pour configurer les affichages.

```python
from __future__ import unicode_literals
from  matplotlib import rcParams
```

*Un exemple d'utilisation*

```python
from math import sin, cos,pi
import numpy as np
rcParams['text.usetex'] = True
rcParams['text.latex.unicode'] = True
x = np.arange(-4*pi, 4*pi, 0.1)
y = 3*np.sin(x+pi/2)
z = 2*np.cos(x+pi/4)
rep = plt.gca(ymargin = 0.1,anchor = 'E',title ='Personnalisation', fc ='#e9f9e9',\
              xlim = (-10,10), ylim = (-5,5), frame_on = True, \
              xbound =(-20,20),ybound = (-6,6),xlabel = '',\
              ylabel = '',\
              xticks = (np.arange(-5*pi, 5*pi, step = pi)),\
              yticks = (np.arange(-6, 6, step = 1)),visible = True )# Récupère le repère actuel ou en crée un
# Les tracés : Ligne avec les légendes en utilisant LaTeX
rep.plot(x,y, label =r'$x \mapsto 3sin(x+\frac{\pi}{2})$', linewidth = 2)
rep.plot(x,z, label =r'$x \mapsto 2cos(x+\frac{\pi}{4})$', linewidth = 2)
# Informations diverses
plt.title('Fonctions trigonométriques',fontname ='Times new roman', size = 16, color ='red')
rep.legend(loc='lower center', bbox_to_anchor=(0.5, -0.15), shadow=True, fontsize= 14, facecolor ='#12ffff' )
# Configurer l'affichage des unités sur l'axe des abscisses
xticks = (np.arange(-5*pi, 5*pi, step = pi))
rep.xaxis.set_major_formatter(plt.FuncFormatter('{:.2f}'.format))
# Afficher la grille
plt.grid(linestyle = '--')
# Centrer les axes
rep.spines['right'].set_visible(False)
rep.spines['top'].set_visible(False)
rep.spines['bottom'].set_position('zero')
rep.spines['left'].set_position('zero')
# Affichage
plt.show() 
```

### C) Utiliser un fichier csv

<p>Avec **matplotlib** on peut aussi représenter des données extraites d'un fichier au format *csv* (Comma-Separated Values).</p>
Le code suivant permet la création d'un fichier au format *csv*, que l'on sauvegarde sur le disque sous le nom de *fichierCSV1*.

```python
from random import randint
fichier = open('fichierCSV1','w')
for i in range(30):
    nom = str(i)
    note = str(randint(0,20))
    fichier.write(nom+','+note+'\n')
fichier.close()
```

La fonction *loadtxt* de ***numpy** permet le chargement du contenu du fichier sous la forme d'un tableau. On peut ensuite représenter les données, par exemple à l'aide de la méthode *bar* du module **pyplot**.

```python
x,y = np.loadtxt('fichierCSV1',unpack=True, delimiter = ',')

plt.bar(x,y)

plt.title('Relevé de notes')
plt.xlabel('Numéro candidat')
plt.ylabel('Notes')

plt.show()
```
